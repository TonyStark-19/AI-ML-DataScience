## Handling the outliers

```python
# drop non-numeric columns
df.head()
numeric_data = df.drop(["city", "country", "name", "region"], axis  = 1)
numeric_data.count()
```

```python
# 25%
Q1 = numeric_data.quantile(0.25)
# 75%
Q3 = numeric_data.quantile(0.75)
# inter quartile range
IQR = Q3 - Q1
IQR
```

```python
# data < (Q1 - 1.5*IQR) → flags lower outliers
# data > (Q3 + 1.5*IQR) → flags upper outliers
# .any(axis=1) → marks a row if any column has an outlier
# ~(...) → negation, keeps only rows without outliers.
data_without_outliers = numeric_data[~((numeric_data < (Q1 - 1.5 * IQR)) | (numeric_data > (Q3 + 1.5 * IQR))).any(axis = 1)]
data_without_outliers.count()
```
