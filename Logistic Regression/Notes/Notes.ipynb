## Logistic Regression ‚Äì Complete Notes

### 1. What is Logistic Regression?

Logistic Regression is a supervised learning algorithm used for classification problems, mainly binary classification.

- Output is categorical
- Predicts probability of a class
- Uses logistic (sigmoid) function instead of a straight line

> Despite the name regression, it is a classification algorithm.

### 2. When to Use Logistic Regression?

- Binary classification
- Probability-based predictions
- Linearly separable or near-linear data
- Interpretable model required

**Examples:**
- Spam vs Not Spam
- Pass vs Fail
- Disease vs No Disease
- Fraud vs Legitimate

### 3. Types of Logistic Regression
 - **Binary Logistic Regression**
    - Used when there are **two classes (0/1)**
  - **Multinomial Logistic Regression**
    - Used when there are **more than two classes**
    - Classes have **no inherent order**
  - **Ordinal Logistic Regression**
    - Used when classes are **ordered or ranked**

### 4. Difference: Linear vs Logistic Regression
- **Output**
  - Linear Regression: Continuous values
  - - Logistic Regression: Probability between **0 and 1**
- **Function**
  - Linear Regression: Straight line
  - Logistic Regression: Sigmoid (S-shaped) curve
- **Loss Function**
  - Linear Regression: Mean Squared Error (MSE)
  - Logistic Regression: Log Loss (Binary Cross-Entropy)
- **Task Type**
  - Linear Regression: Regression
  - Logistic Regression: Classification

### 7. Decision Boundary

- Threshold usually = 0.5
- If probability ‚â• 0.5 ‚Üí Class 1
- Else ‚Üí Class 0

$\hat{y} = \begin{cases} 1, & \text{if } P \ge 0.5 \\ 0, & \text{otherwise.} \end{cases}$

### 8. Cost Function (Log Loss)

Linear regression uses MSE ‚Üí bad for classification
Logistic regression uses Log Loss / Binary Cross Entropy

`Loss Function:`

$J = - \frac{1}{N} \sum_{i=1}^N [y_i \log(p_i) + (1 - y_i) \log(1 - p_i)]$

> Penalizes confident wrong predictions heavily

### 11. Assumptions of Logistic Regression
- Binary dependent variable
- Independent observations
- No multicollinearity
- Linear relationship between features & log-odds
- Large sample size preferred

### 12. Evaluation Metrics
Accuracy alone is not enough
- **Evaluation Metrics**
  - **Accuracy**
    - Formula: (TP + TN) / Total
    - Meaning: Overall correctness of the model
  - **Precision**
    - Formula: TP / (TP + FP)
    - Meaning: How many predicted positives are actually correct
  - **Recall**
    - Formula: TP / (TP + FN)
    - Meaning: How many actual positives are correctly identified
  - **F1-Score**
    - Formula: 2PR / (P + R)
    - Meaning: Balance between precision and recall
  - **ROC-AUC**
    - Formula: Area under the ROC curve
    - Meaning: Model‚Äôs ability to distinguish between classes

### 13. Confusion Matrix
- **Actual Class: 0**
    - Predicted 0 ‚Üí True Negative (TN)
    - Predicted 1 ‚Üí False Positive (FP)
- **Actual Class: 1**
    - Predicted 0 ‚Üí False Negative (FN)
    - Predicted 1 ‚Üí True Positive (TP)

### 14. Advantages
- Simple & fast
- Interpretable coefficients
- Probabilistic output
- Works well with small datasets

### 15. Disadvantages
- Assumes linear decision boundary
- Sensitive to outliers
- Poor performance on complex data
- Needs feature engineering

### 16. Hyperparameters
- **C**
    - Meaning: Inverse of regularization strength  
    - Smaller value ‚Üí stronger regularization
- **penalty**
    - Meaning: Type of regularization applied
    - Common values: `l1`, `l2`
- **solver**
    - Meaning: Optimization algorithm used to train the model
    - Common solvers: `liblinear`, `lbfgs`, `saga`
- **max_iter**
    - Meaning: Maximum number of iterations allowed for convergence
