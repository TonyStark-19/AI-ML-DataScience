### Load the libraries needed

```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import warnings
warnings.filterwarnings("ignore")
```

### Load the Data

```python
df = pd.read_csv("insurance_data.csv")
df.head()
```

### Split the data into **dependent** and **independent** variables

```python
# independent
x = df[["age"]]
# dependent
y = df.bought_insurance
```

```python
x.head()
```

```python
y.head()
```

### Now visulise the data to find the pattern

```python
# Plot a scatter chart
plt.title("Will buy the insurance or not")
plt.scatter(x, y)
plt.xlabel("Age values")
plt.ylabel("Will buy or not")
plt.tight_layout()
plt.show()
```

```python
# Import Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
# Create object
model = LogisticRegression()
```

```python
# Pass the values to train the model
model.fit(x, y)
```

```python
predicted_y = model.predict(x)
predicted_y
```

```python
# Plot a predicted value scatter chart
plt.title("Will buy the insurance or not")
plt.scatter(x, y)
plt.xlabel("Age values")
plt.ylabel("Will buy or not")
plt.plot(x, predicted_y)
plt.tight_layout()
plt.show()
```

```python
# Make the User enter the area values
age = int(input("Enter the value of age :"))
# indexing to reduce the dimension
if model.predict([[age]])[0] == 0:
    print("Person will not buy the insurance")
else:
    print("Person will buy the insurance")
```

```python
# Convert model into a pickle file
pickle.dump(model, open("model3.pkl", "wb"))
```
