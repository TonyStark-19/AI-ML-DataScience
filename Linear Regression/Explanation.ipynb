```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# data
weeks_data = [1, 2, 3, 4, 5]
sales_data = [2, 4, 5, 4, 5]

# convert data to dataframe
df = pd.DataFrame({"Weeks": weeks_data,
                   "Sales": sales_data})
df
```

```python
# differentiate data in independent(feature) and dependent(target) variables
X = df["Weeks"].values
Y = df["Sales"].values
```

```python
# calculate mean
mean_x = np.mean(X)
mean_y = np.mean(Y)

print("Mean of x :", mean_x)
print("Mean of y :", mean_y)
```

```python
# formula y = mx + c
n = len(X)

# numerator and denominator
numer = 0
denom = 0

# calculate numerator and denominator
for i in range(n):
    # x - x bar * y - y bar
    numer += (X[i] - mean_x) * (Y[i] - mean_y) 
    # x - x bar whole square
    denom += (X[i] - mean_x) ** 2

# m (slope of the line)
m = numer/denom
# c (intercept of line)
c = mean_y - (m * mean_x)

print("m:", m, "c:", c)
```

```python
# max and min values
max_x = np.max(X) + 1
min_x = np.min(X) - 1

# plotting variables for regression line
x = np.linspace(min_x, max_x)
y = (m * x) + c

# plot line chart
plt.plot(x,y,color = "#8c250f", label = "Regression line", linestyle = "--")
plt.scatter(X,Y,color = "#040378", label = "Data points", s = 60)
plt.title("Linear Regression")
plt.xlabel("Weeks")
plt.ylabel("Sales")
plt.grid(alpha = 0.6)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# R2
# sum of squared total
ss_tot  = 0
# sum of squared residual
ss_res = 0

for i in range(n):
    y_pred = (m * X[i]) + c
    ss_tot += (Y[i] - mean_y) **2
    ss_res += (Y[i] - y_pred) **2

r2 = 1 - (ss_res/ss_tot)
print("R2:", r2)
```

```python
# validate
# import linear regression
from sklearn.linear_model import LinearRegression
```

```python
# reshape X to 2D
X = X.reshape((n, 1))
X
```

```python
# create object
model = LinearRegression()
```

```python
# pass values for training
model.fit(X, Y)
```

```python
# R2
model.score(X, Y)
```
