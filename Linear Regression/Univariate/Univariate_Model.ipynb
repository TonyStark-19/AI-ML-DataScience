### Load the libraries needed

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

### Load the Data

```python
df = pd.read_csv("areaprice071.csv")
df
```

### Split the data into **dependent** and **independent** variables

```python
# Independent
x = df[["area"]]
# dependent
y = df.price
```

```python
x
```

```python
y
```

### Now visulise the data to find the pattern

```python
# Plot scatter chart
plt.title("Area vs price")
plt.scatter(x, y, color="red")
plt.xlabel("Area values")
plt.ylabel("Price values")
plt.tight_layout()
plt.show()
```

```python
# Import linear regression
from sklearn.linear_model import LinearRegression
```

```python
# Create object
model = LinearRegression()
```

```python
# Pass the values to train the model
# Note: fit function takes first value as 2D
model.fit(x, y)
```

```python
# Coefficient value m in ( y = mx + c )
m = model.coef_
```

```python
# Intercept value c in ( y = mx + c )
c = model.intercept_
```

```python
# Predict the value from the model ( val * m + c )
predicted_y = model.predict(x)
predicted_y
```

```python
# Plot predicted value chart
plt.title("Area vs price")
# original data
plt.scatter(x, y, color="red")
plt.xlabel("Area values")
plt.ylabel("Price values")
# predicted data
plt.plot(x, predicted_y, color="green")
plt.tight_layout()
plt.show()
```

```python
# Make the User enter the area values
area1 = int(input("Enter the value of area :"))
# indexing to reduce the dimension
price1 = model.predict([[area1]])[0].round()
print("Value of price", area1, "sq feet is :", int(price1), "rupees")
```
